```python
import urllib.request
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import re
import json

data = pd.read_csv('url_technology.csv')
title = []
sub_titles = []
articles = []
author=[]
author_urls = [] 
claps = []
reading_time = []
image_sources = []
data_urls_scraped = []

for i in range(600):
    try:
        print(i)
        url_to_scrap = data.iloc[i]['url']
        r = requests.get(url_to_scrap, timeout=20)    
        #get all the data concatenated
        text_got = r.text+''
        r.close
        
        data_urls_scraped.append(url_to_scrap)
        soup = BeautifulSoup(text_got, "html.parser")
        headings = []
        sub_headings = []
        # Heading
        for my_tag in soup.find_all('h1'):
            headings.append(my_tag.text)
        if(len(headings) > 0):
            title.append(headings[0])
        # Subheading
        for my_tag1 in soup.find_all('h2'):
            sub_headings.append(my_tag1.text)
        if(len(sub_headings) > 0):
            sub_titles.append(sub_headings[0])
        para_article = ""
        # Article
        for my_tag2 in soup.find_all('p'):
            para_article = para_article + (my_tag2.text)
        articles.append(para_article)
        # author
        script_tag1 = soup.find('script', type='application/ld+json')
        if script_tag1:
            # Extract the JSON data
            json_data1 = json.loads(script_tag1.string)
            # Extract the author's name
            author_name = json_data1['author']['name']
        author.append(author_name)
        
        #Author_link
        script_tag11 = soup.find('script', type='application/ld+json')
        if script_tag11:
            # Extract the JSON data
            json_data11 = json.loads(script_tag11.string)
            # Extract the author's name
            author_url = json_data11['author']['url']
        author_urls.append(author_url)
        
        #reading_time = []
        R_div = soup.find('div', class_='ab ae')
        reading_time.append((R_div.find('span')).text)
        
        #cLAPS 
        script_tag = soup.find('script', string=re.compile(r'window\.__APOLLO_STATE__'))
        if script_tag:
            # Extract the JavaScript code inside the script tag
            javascript_code = script_tag.string
            # Use regex to find the clap count
            clap_count_match = re.search(r'"clapCount":(\d+)', javascript_code)
            if clap_count_match:
                clap_count = int(clap_count_match.group(1))
                claps.append(clap_count)
            else:
                print("Clap count not found in code.")
        
        # Image sources
        for my_tag4 in soup.find_all('img'):
            image_sources.append(my_tag4.get('src'))
            
    except:
        print("no data found")
```

```python
# Combine all lists into a list of tuples
data = list(zip(data_urls_scraped,title, sub_titles, author, author_urls, claps, reading_time, image_sources,articles))
# Create a DataFrame from the list of tuples
df = pd.DataFrame(data, columns=["Url Scrapped",'Title', 'Subtitles', 'Author', 'Author URLs', 'Claps', 'Reading Time', 'Image Sources','Articles'])
# Write the DataFrame to a CSV file
df.to_csv('scraped_data.csv', index=False)
```
